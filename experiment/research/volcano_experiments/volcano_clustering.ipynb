```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from psych_metric.datasets.pa.dataset import PA
from psych_metric.datasets.aflw.dataset import AFLW

from psych_metric.metrics.volcano.metric import VolcanoMetricMultinomialEM, VolcanoMetricMultinomialMC
```

# Cluster First Impressions Data

```python
ds = AFLW(stage='test', augment=False, numpy=False)
X = ds.get_multinomial_arrays()['Trustworthiness']
```

```python
for n_classes in range(7,8):
    for trial in range(1):
        Metric = VolcanoMetricMultinomialEM(n_classes=n_classes)
        Metric.train(X, verbose=False)
        Metric.plot_prior()
        Metric.plot_posterior()
```
